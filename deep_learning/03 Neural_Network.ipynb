# 神经网络

## 激活函数

### sigmoid 函数

$$
h(x) = {1 \over 1 + e^{-x}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, sigmoid(x))
plt.show()
```

### 阶跃函数

```python
def step_function(x: np.ndarray):
    y = x > 0
    return y.astype(np.int32)
```

```python
plt.plot(x, sigmoid(x), label='sigmoid')
plt.plot(x, step_function(x), linestyle="--", label='step')
plt.legend()
plt.show()
```

## ReLU 函数

$$
h(x) = \begin{cases}
x & (x > 0) \\
0 & (x \leqslant 0)
\end{cases}
$$

```python
def relu(x):
    return np.maximum(0, x)
```

```python
plt.plot(x, sigmoid(x), label='sigmoid')
plt.plot(x, step_function(x), linestyle="--", label='step')
plt.plot(x, relu(x), linestyle="-.", label='step')
plt.legend()
plt.ylim(-0.1, 2)
plt.show()
```

```python
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a - c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```
