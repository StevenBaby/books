## matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)
```

```python
plt.plot(x, y)
plt.show()
```

### plt 的功能

```python
y2 = np.cos(x)

plt.plot(x, y, label="sin(x)")
plt.plot(x, y2, label="cos(x)", linestyle='--')
plt.xlabel("x")
plt.ylabel("y")
plt.title(r"$\sin(x)$ & $\cos(x)$")
plt.legend()
plt.show()
```
